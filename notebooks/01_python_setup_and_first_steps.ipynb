## Section 2 — Recommended Local Setup

We recommend using **Anaconda** (or Miniconda) to manage Python and packages.  
A dedicated **conda environment** keeps course libraries isolated from other projects.

### Why a separate environment?

Different projects may need different library versions.  
A dedicated environment prevents version conflicts and keeps things reproducible.

Our environment will be called **`finance_ml`** (you can choose another name).

### 2.1 Windows setup

Open **Anaconda Prompt** (search for it in the Start menu) and run the following commands  
one at a time. **Do not run these inside the notebook** — they are terminal commands.

```

Open **Terminal** and run:

```
conda create -n finance_ml python=3.11 -y
conda activate finance_ml
pip install pandas numpy matplotlib scikit-learn xgboost jupyter ipykernel

## Section 4 — Kernel Selection and Common Issues

### What is a kernel?

The **kernel** is the Python process running behind your notebook.  
When you press *Shift+Enter*, the cell's code is sent to the kernel for execution.

The kernel determines **which Python installation** (and which packages) are available.

### How to select the right kernel

- **Jupyter Notebook / JupyterLab:** look at the top-right corner for the kernel name.
- **VS Code:** click the kernel name in the top-right of the notebook editor.
- **Colab:** the kernel is managed automatically.

### Common symptoms of a wrong kernel

| Symptom | Likely cause |
|---------|-------------|
| `ModuleNotFoundError: No module named 'pandas'` | Wrong kernel or package not installed in this environment |
| Notebook runs in Colab but fails locally | Missing local packages |
| `pip install` succeeds but import still fails | Installed in a different environment |

### Troubleshooting checklist

1. Did I **activate** the correct environment? (`conda activate finance_ml`)
2. Did I **install** the library inside that environment?
3. Did I **register** the kernel? (`python -m ipykernel install ...`)
4. Did I **select** the `finance_ml` kernel in Jupyter?
5. Did I **restart the kernel** after installation? (Kernel → Restart)

## Section 5 — First Python Cells

Let's write a few lines of Python to make sure everything works.  
None of this is finance-specific yet — it is just to get comfortable.

````python
# Variables and arithmetic
revenue = 150_000
costs = 87_500
profit = revenue - costs

print("Revenue:", revenue)
print("Costs:  ", costs)
print("Profit: ", profit)
````

````python
# Strings
company_name = "Acme Finance"
greeting = f"Welcome to {company_name}!"
print(greeting)
````

````python
# A list of values
monthly_returns = [0.02, -0.01, 0.03, 0.015, -0.005, 0.04]
print("Monthly returns:", monthly_returns)
print("Number of months:", len(monthly_returns))
print("First return:", monthly_returns[0])
````

````python
# A dictionary (key-value pairs)
portfolio = {
    "equities": 0.60,
    "bonds": 0.25,
    "cash": 0.10,
    "alternatives": 0.05,
}

for asset, weight in portfolio.items():
    print(f"{asset:15s} {weight:.0%}")
````

````python
# A quick calculation
total_weight = sum(portfolio.values())
print(f"Total portfolio weight: {total_weight:.0%}")
````

## Section 6 — Your First DataFrame

A **DataFrame** is simply a table — rows and columns, like a spreadsheet.  
pandas is the library that gives us DataFrames in Python.

Let's create a tiny one from scratch:

````python
import pandas as pd

data = {
    "client_id": ["C001", "C002", "C003", "C004", "C005"],
    "monthly_income": [3200, 4800, 2900, 5500, 3700],
    "product_type": ["Mortgage", "Savings", "Loan", "Savings", "Mortgage"],
    "default_flag": [0, 0, 1, 0, 0],
}

df = pd.DataFrame(data)
df
````

````python
# Access a single column
df["monthly_income"]
````

````python
# Calculate the mean income
mean_income = df["monthly_income"].mean()
print(f"Average monthly income: {mean_income:,.0f} EUR")
````

````python
# Quick summary statistics
df.describe()
````

> **Takeaway:** a DataFrame is just a table, and you can already interact with it —  
> select columns, compute statistics, and display results. We will do much more in Notebook 02.
